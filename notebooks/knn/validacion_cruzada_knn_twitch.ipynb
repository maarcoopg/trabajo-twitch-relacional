# 05 - kNN: validación cruzada

En este notebook se evalúa el modelo **kNN** mediante validación cruzada.

La validación cruzada ayuda a comprobar si el rendimiento del modelo es consistente en distintas particiones del dataset.

## Objetivo metodologico

La validacion cruzada de kNN evalua si el rendimiento observado es estable. Es especialmente relevante en kNN porque el resultado depende mucho de la distribucion local de ejemplos en el espacio de variables.


## 1. Importación de librerías y configuración inicial

Se importan las funciones necesarias y se definen las rutas del proyecto.

```python
from pathlib import Path
import sys

import pandas as pd
from sklearn.model_selection import cross_val_score

sys.path.append("../../")

from src.data_loader import load_processed_dataset
from src.model_training import get_base_models

ROOT = Path("../../").resolve()
DATA_PROCESSED = ROOT / "data" / "processed"
RESULTS_DIR = ROOT / "results"

RESULTS_DIR.mkdir(parents=True, exist_ok=True)
```

## 2. Validación cruzada del modelo

Se evalúa kNN con 5 particiones y usando `f1` como métrica principal.

### Lectura de resultados

Una `std_f1` baja indica que el modelo se comporta de forma parecida en distintas particiones. Si la desviacion es alta, conviene revisar escalado, hiperparametros o el conjunto de variables.


```python
df = load_processed_dataset(DATA_PROCESSED / "twitch_mature_features.csv")
X = df.drop(columns=["new_id", "mature"])
y = df["mature"]
model = get_base_models(random_state=42)["knn"]
scores = cross_val_score(model, X, y, scoring="f1", cv=5, n_jobs=-1)
cv_metrics = pd.DataFrame([{
    "model": "knn",
    "mean_f1": scores.mean(),
    "std_f1": scores.std(),
    "fold_scores": ",".join(f"{score:.4f}" for score in scores)
}])
cv_metrics
```

## 3. Guardado de métricas de validación cruzada

Se guardan los resultados para la comparativa final.

```python
cv_metrics.to_csv(RESULTS_DIR / "knn_cv_metrics.csv", index=False)
```

## 4. Conclusiones

La validación cruzada permite comprobar la estabilidad del modelo kNN y comparar su rendimiento con el resto de modelos.
